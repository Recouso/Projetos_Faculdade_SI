```python
# Criação da classe Simplex para o funcionamento do cálculo Max
class Simplex:

    # Metodo que define o construtor que vai iniciar a tabela
    def __init__(self):
        self.tabela = []

    # Metodo que armazena a definicao da funcao objetivo(fo)
    def objetivo_funcao(self,fo:list):
        self.tabela.append(fo)

    # Metodo que armazena a definicao do "sujeito a"(sa)
    def adicionar_restricoes(self,sa:list):
        self.tabela.append(sa)

    # Metodo que retorna a coluna que vai entrar, retornando um inteiro(indice da coluna)
    def get_entrada_coluna(self) -> int:
        coluna_pivo = min(self.tabela[0]) # Retorna o menor valor da fun��o objetivo
        index = self.tabela[0].index(coluna_pivo)
        return index

    # Metodo que identifica a linha que sai
    def get_linha_saida(self,entrada_coluna: int) -> int:
        resultados = {}
        for linha in range(len(self.tabela)):
            if linha > 0:
                if self.tabela[linha][entrada_coluna] > 0:
                    divisao = self.tabela[linha][-1] / self.tabela[linha][entrada_coluna]
                    resultados[linha] = divisao
        index = min(resultados, key=resultados.get)
        return index

    # Metodo que calcula e retorna a nova linha do pivo
    def calcular_nova_linha_pivo(self,entrada_coluna: int, exit_linha: int) -> list:
        linha = self.tabela[exit_linha]
        pivo = linha[entrada_coluna]
        nova_linha_pivo = [valor / pivo for valor in linha]
        return nova_linha_pivo

    # Metodo que calcula e retorna a nova linha
    def calcular_nova_linha(self,linha: list, entrada_coluna: int, linha_pivo: list) -> list:
        pivo = linha[entrada_coluna] * -1
        linha_resultado = [valor * pivo for valor in linha_pivo]
        nova_linha = []

        for i in range(len(linha_resultado)):
            soma_valor = linha_resultado[i] + linha[i]
            nova_linha.append(soma_valor)
        return nova_linha

    # Metodo que enquanto a primeira linha(fo) conter numeros negativos, a tabela tera que ser recalculada
    def e_negativo(self) -> bool:
        negativo = list(filter(lambda x: x < 0,self.tabela[0]))
        return True if len(negativo) > 0 else False

    # Metodo que imprime a tabela final
    def mostrar_tabela(self):
        print(" x0 \t x1 \t x2 \t x3 \t x4 \t x5 \tSolucao")
        for i in range(len(self.tabela)):
            for j in range(len(self.tabela[0])):
                print('{:.2f}\t'.format(self.tabela[i][j]),end="")
            print()
        print('\nSolucao: {:.2f}\t'.format(self.tabela[0][j]))
        print('x0: {:.2f}\t'.format(self.tabela[1][j]))
        print('x1: {:.2f}\t'.format(self.tabela[3][j]))

    # Metodo que recalcula a tabela
    def recalcular(self):
        entrada_coluna = self.get_entrada_coluna()
        primeira_linha_saida = self.get_linha_saida(entrada_coluna)
        linha_pivo = self.calcular_nova_linha_pivo(entrada_coluna,primeira_linha_saida)
        self.tabela[primeira_linha_saida] = linha_pivo
        copia_tabela = self.tabela.copy()
        index = 0

        while index < len(self.tabela):
            if index != primeira_linha_saida:
                linha = copia_tabela[index]
                nova_linha = self.calcular_nova_linha(linha,entrada_coluna,linha_pivo)
                self.tabela[index] = nova_linha
            index += 1

    # Metodo que enquanto tiver numero negativo, ele ira chamar o metodo e calcular de novo
    # ate que se gere uma tabela que tenha a primeira linha como numeros positivos
    def resolver(self):
        self.recalcular()
        while self.e_negativo():
            self.recalcular()
        self.mostrar_tabela()
```

```python
# Primeira entrada de dados do programa 
qtdVar = 21
qtdRes = 21
tipo = "a"
coeficientes = []
todas_restricoes = []
while (qtdVar > 20):
  qtdVar = int(input('\nQuantas variaveis (maximo 20)? '))

while (qtdRes > 20):
  qtdRes = int(input('Quantas restricoes (maximo 20)? '))

print("")

tipo = str(input('Tipo de problema (Max/Min): '))

print("")
print("Funcao Objetivo")
for x in range(qtdVar):
  coeficientes.append(int(input("Coeficiente da variavel x{}: ".format(x))))

for x in range(qtdRes):
  print("")
  restricoes = []
  variaveisAux = []
  print("Restricao {}: ".format(x+1))
  for y in range(qtdVar):
    variaveisAux.append(int(input("Coeficiente da variavel x{}: ".format(y))))

  restricoes.append(str(input("Tipo de Restrição: ")))
  restricoes.append(int(input("Termo independente: ")))

  for z in range(qtdVar):
    restricoes.append(variaveisAux[z])

  todas_restricoes.append(restricoes)

valores_tabela = []
valor_restricao = []
count = 0
if (tipo == 'Max'):
  valor_restricao.append(0)
  for coeficiente in coeficientes:
    valor_restricao.append(coeficiente * -1)
  valores_tabela.append(valor_restricao)
  for restricao in todas_restricoes:
    valor_restricao = []
    if (restricao[0] == '>='):
      count += 1
      for x in range(qtdVar + 1):
        valor_restricao.append(restricao[x+1] * -1)
      valores_tabela.append(valor_restricao)
    elif (restricao[0] == '<='):
      count += 1
      for x in range(qtdVar + 1):
        valor_restricao.append(restricao[x+1])
      valores_tabela.append(valor_restricao)
    elif (restricao[0] == '=='):
      count += 2
      for x in range(qtdVar + 1):
        valor_restricao.append(restricao[x+1])
      valores_tabela.append(valor_restricao)
      valor_restricao = []
      for x in range(qtdVar + 1):
        valor_restricao.append(restricao[x+1] * -1)
      valores_tabela.append(valor_restricao)
  i = 0
  for array in valores_tabela:
    if (i == 0):
      for x in range(count):
        array.append(0)
    else:
      for x in range(count):
        if (x == i - 1):
          array.append(1)
        else:
          array.append(0)
    i += 1
elif (tipo == 'Min'):
  count = 0
  func_obj = []
  var_each_line = []
  termos_independentes = []
  for coeficiente in coeficientes:
    func_obj.append(coeficiente)

  for restricao in todas_restricoes:
    valor_restricao = []
    if (restricao[0] == '>='):
      count += 1
      for x in range(1,len(restricao)):
        if x == 1:
          termos_independentes.append(restricao[x] * -1)
        else:
          valor_restricao.append(restricao[x] * -1)
      var_each_line.append(valor_restricao)
    elif (restricao[0] == '<='):
      count += 1
      for x in range(1,len(restricao)):
        if x == 1:
          termos_independentes.append(restricao[x])
        else:
          valor_restricao.append(restricao[x])
      var_each_line.append(valor_restricao)
    elif (restricao[0] == '=='):
      count += 2
      for x in range(1,len(restricao)):
        if x == 1:
          termos_independentes.append(restricao[x])
        else:
          valor_restricao.append(restricao[x])
      var_each_line.append(valor_restricao)

      valor_restricao = []
      for x in range(1,len(restricao)):
        if x == 1:
          termos_independentes.append(restricao[x] * -1)
        else:
          valor_restricao.append(restricao[x] * -1)
      var_each_line.append(valor_restricao)
```

```python
# Printar MAX ou Printar MIN
if tipo == 'Max':
  for array in valores_tabela:
    var_aux = array[0]
    del array[0]
    array.append(var_aux)

  simplex = Simplex()
  i = 0
  for array in valores_tabela:
    if i == 0:
      simplex.objetivo_funcao(array)
    else:
      simplex.adicionar_restricoes(array)
    i += 1
  print()
  print("Maximização (Quadro Final): \n")
  simplex.resolver()
elif tipo == 'Min':
  print("\nMinimização: \n")
  print(optimize.linprog(
      c = func_obj,
      A_ub = var_each_line,
      b_ub = termos_independentes,
      bounds=(None, None),
      method='simplex'
  ))
  print("\n**** OBS: fun = Solução ****\n**** x = x0 e x1 respectivamente ****")
```
